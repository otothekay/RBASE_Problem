```python
import numpy as np

aircraft = 24

HAS = range(25)
PAS = range(25)
SUN = range(25)
DECOY = range(25)


table = []

for h in HAS:
    for p in PAS:
        for s in SUN:
            for d in DECOY:
                row = []
                row.append(h)
                row.append(p)
                row.append(s)
                Opn = max(0, 24-h-p-s)
                row.append(Opn)
                row.append(d)
                table.append(row)

table           
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(table, columns=['HAS','PAS','Sunshades','Open','Decoys'])
df['cost'] = 7.8*df['HAS'] + 3.6*df['PAS'] + 0.4*df['Sunshades'] + 0.1*df['Decoys']
df
```

```python
p_hit_HAS, p_hit_PAS, p_hit_Sun, p_hit_Open = 0.2, 0.5, 1, 1

df['P_damage'] = (p_hit_HAS*df['HAS'] + p_hit_PAS*df['PAS'] + p_hit_Sun*df['Sunshades'] + p_hit_Open*df['Open']) / (df['HAS'] + df['PAS'] + df['Sunshades'] + df['Open'])
df
```

```python
df['P_ID'] = aircraft / (df['HAS']+df['PAS']+df['Sunshades']+df['Open']+df['Decoys']) 
df['P_k'] = df['P_damage'] * df['P_ID']
df['delta_P_K'] = 1 - df['P_k']
df['Cost_Effectiveness'] = df['delta_P_K'] / df['cost']
df
```

```python
sorted_df = df.sort_values(by=['Cost_Effectiveness','cost'], ascending=False, ignore_index=True)
sorted_df[0:24]
```

```python
import plotly.express as px

fig = px.scatter(sorted_df[0:500], y='Cost_Effectiveness', x='cost', color='Decoys')
fig.show()
```
